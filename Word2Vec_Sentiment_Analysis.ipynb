## Clean Dataset

### Imports

```python
from nltk.corpus import stopwords
from bs4 import BeautifulSoup 
import re

nltk.download('stopwords')
```

### A function for cleaning reviews

```python
# function to clean and pre-process the text.
def clean_reviews(review):  
    
    # 1. Removing html tags
    review_text = BeautifulSoup(review,"lxml").get_text()
    
    # 2. drop numbers
    review_text = re.sub("[^a-zA-Z]"," ",review_text)
    
    # 3. Converting to lower case and splitting
    review_text = review_text.lower().split()
    
    # 4. Remove stopwords
    stop_words = set(stopwords.words("english"))     
    word_tokens = [word for word in review_text if not word in stop_words]
    
    cleaned_review = " ".join(word_tokens)

    return cleaned_review
```

### Preprocessing 
Here we limit the dataset to only the necessary columns, generate our target column and clean the reviews.

As explained in previous milestones, the target column is generated by interpretting the score column.  If the score is > 3, then we return a 1, otherwise 0.  This is our 'Sentiment'.

We also balance the dataset by subsampling 50,000 examples from each sentiment.

```python
df = df[['Text','Score']]
#print(df.shape)

df.drop_duplicates(subset=['Text','Score'],keep='first',inplace=True)

df['sentiment'] = df['Score'].apply(lambda x: 1 if x>3 else 0)

# subsample so we have even distribution
pos_sent = df.loc[df.sentiment==1,:]
neg_sent = df.loc[df.sentiment==0,:]

pos_rand_smpl = pos_sent.sample(n=50000, replace=False, random_state=1)
neg_rand_smpl = neg_sent.sample(n=50000, replace=False, random_state=1)

# join back together
train_df = pd.concat([pos_rand_smpl, neg_rand_smpl], ignore_index=True)
```

```python
# get cleaned review text string
train_df['clean_review']=train_df['Text'].apply(clean_reviews)
# make it a list of tokens for Word2Vec
train_df['words'] = train_df['clean_review'].apply(lambda x: x.split())
```

### Visualization Function
Below is a function used to visualize the vectors for the words in our dataset.

This function was adapted from:
https://www.kaggle.com/code/jeffd23/visualizing-word-vectors-with-t-sne/notebook

```python
def tsne_plot(model, neighbor=False):
  labels = []
  tokens = []
  # introduce a counter so that we do not plot the entire dataset
  counter = 0

  # NOTE
  # code block bleow allowed me to generate a plot of 20 words, and their 5 closest neighbors in the vocabulary
  if neighbor:
    for word in list(model.wv.key_to_index.keys()):
      tokens.append(model.wv[word])
      labels.append(word)
      for similar_words in list(w2v.wv.most_similar(word, topn=5)):
        tokens.append(model.wv[similar_words[0]])
        labels.append(similar_words[0])
      
      if counter >= 20:
        break
      
      counter = counter + 1
  # else, we simply plot 1000 words
  else:
    for word in list(model.wv.key_to_index.keys()):
      tokens.append(model.wv[word])
      labels.append(word)

      if counter >= 500:
        break
      
      counter = counter + 1
  
  tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
  new_values = tsne_model.fit_transform(np.asarray(tokens))

  x = []
  y = []
  for value in new_values:
    x.append(value[0])
    y.append(value[1])
    
  plt.figure(figsize=(10, 10)) 
  for i in range(len(x)):
    plt.scatter(x[i],y[i])
    plt.annotate(labels[i],
                  xy=(x[i], y[i]),
                  xytext=(5, 2),
                  textcoords='offset points',
                  ha='right',
                  va='bottom')
  plt.show()
```

## Sentiment Analysis

### Imports

```python
#keras
import keras
from keras.preprocessing.text import one_hot,Tokenizer
from keras.utils import pad_sequences
from keras.models import Sequential
from keras.layers import Dense,Flatten,Embedding,Input,CuDNNLSTM,LSTM,Bidirectional,GlobalMaxPool1D
from keras.models import Model
from keras.preprocessing.text import text_to_word_sequence

from keras.initializers import Constant
from keras.layers import ReLU
from keras.layers import Dropout

from sklearn.model_selection import train_test_split,cross_validate
from sklearn.metrics import confusion_matrix
```

### Final text preprocessing steps

here we will tokenize all of our sentences, and pad them out so that they are all the same length.  We will also generate our embedding matrix.

```python
# get max length sentence
measurer = np.vectorize(len)
max_len = measurer(train_df.words).max(axis=0)

print(max_len)
```

```python
# tokenize the text
tok = Tokenizer()
tok.fit_on_texts(train_df['clean_review'])

vocab_size = len(tok.word_index) + 1

train_df['tok'] = tok.texts_to_sequences(train_df['clean_review'])
```

```python
# pad the reviews
pad_rev = pad_sequences(train_df['tok'], maxlen=max_len, padding='post')

train_df['pad_rev'] = [pad_rev[i] for i in range(pad_rev.shape[0])]
```

```python
word_vec_dict={}
for word in list(w2v.wv.key_to_index.keys()):
  word_vec_dict[word]=w2v.wv.get_vector(word)
```

```python
# Create the embedding matrix
embed_matrix = np.zeros(shape=(vocab_size, 100))
for word,i in tok.word_index.items():
  embed_vector = word_vec_dict.get(word)
  if embed_vector is not None:
    embed_matrix[i]=embed_vector
```

### Build the model

```python
model=Sequential()

model.add(Embedding(input_dim=vocab_size, output_dim=100, input_length=max_len, embeddings_initializer=Constant(embed_matrix)))
model.add(Bidirectional(LSTM(25,return_sequences=True)))
model.add(GlobalMaxPool1D())
model.add(Flatten())
model.add(Dense(16,activation='relu'))
model.add(Dropout(0.50))
model.add(Dense(2,activation='sigmoid'))
```

```python
model.summary()
```

```python
# compile the model
model.compile(optimizer=keras.optimizers.RMSprop(lr=1e-3), loss='binary_crossentropy', metrics=['accuracy'])
```

### Split the data

```python
# prepare train and val sets first
Y = keras.utils.to_categorical(train_df['sentiment'])
x_train, x_test, y_train, y_test = train_test_split(train_df[['Score','pad_rev']], Y, test_size=0.20, random_state=42)

# REMOVE THE SCORES FROM TRAIN
x_train = np.asarray(x_train['pad_rev'])
x_train = np.stack(x_train)
#print(x_train)

# create validation sets
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.20, random_state=42)
```

#### Below we create adversy testing sets

To test the robustness of the model, we are creating more adverse data sets.  

First we create small subsets (1000 examples) of each score (1 through 5), this will allow us to see how the model reacts to very polarizing and some more mild reviews.

Importantly we then drop the 'Score' column from our testing dataset.  This was left in before splitting so that i could properly subsample the testing set.

We also create two datasets that are imbalanced (ie 80% pos, 20% neg).  This will help us mimic the imbalance in the actual dataset.

```python
# create subsets for each rating (score)

# subsample
# 1 star
x_one_star = x_test.loc[x_test.Score==1,:].sample(n=1000, replace=False, random_state=42)
x_one_star = np.stack(np.array(x_one_star['pad_rev']))
# 2 star
x_two_star = x_test.loc[x_test.Score==2,:].sample(n=1000, replace=False, random_state=42)
x_two_star = np.stack(np.array(x_two_star['pad_rev']))
# 3 star
x_three_star = x_test.loc[x_test.Score==3,:].sample(n=1000, replace=False, random_state=42)
x_three_star = np.stack(np.array(x_three_star['pad_rev']))
# 4 star
x_four_star = x_test.loc[x_test.Score==4,:].sample(n=1000, replace=False, random_state=42)
x_four_star = np.stack(np.array(x_four_star['pad_rev']))
# 5 star
x_five_star = x_test.loc[x_test.Score==5,:].sample(n=1000, replace=False, random_state=42)
x_five_star = np.stack(np.array(x_five_star['pad_rev']))

# create target Y
# all zero for 1-3, all 1 for 4-5
Y_low, Y_high = np.zeros((1000,2), dtype=float), np.zeros((1000,2), dtype=float)
Y_low[:,0] = 1.0
Y_high[:,1] = 1.0
```

```python
# VERY IMPORTANT
# DROP EVERY COLUMN EXCEPT FOR THE PADDED REVIEW
#x_train = x_train.reshape((x_train.shape[0], max_len))
x_test = np.array(x_test['pad_rev'])
x_test = np.stack(x_test)
```

```python
# create imbalanced test sets
# get the locations oft he positives and negatives
pos_loc = np.where(y_test.argmax(axis=1)==0)[0]
neg_loc = np.where(y_test.argmax(axis=1)==1)[0]

# now sample 80% of the pos, 20% of the neg and stick them together
pos, neg = np.random.choice(pos_loc, size=int(pos_loc.shape[0] * 0.8), replace=False), np.random.choice(neg_loc, size=int(neg_loc.shape[0] * 0.2), replace=False)
x_test2 = np.concatenate((x_test[pos],x_test[neg]), axis=0)
y_test2 = np.concatenate((y_test[pos],y_test[neg]), axis=0)

# vice versa
pos, neg = np.random.choice(pos_loc, size=int(pos_loc.shape[0] * 0.2), replace=False), np.random.choice(neg_loc, size=int(neg_loc.shape[0] * 0.8), replace=False)
x_test3 = np.concatenate((x_test[pos],x_test[neg]), axis=0)
y_test3 = np.concatenate((y_test[pos],y_test[neg]), axis=0)
```

### Fitting and testing the model

```python
# specify batch size and epocj=hs for training.
epochs = 5
batch_size = 64

# fitting the model.
model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(x_test, y_test))
```

#### Results

Here we test the model on our balanced testing set, and produce a confusion matrix.  As we can see from the results, the model handles the testing set well, and reports an accuracy close to that of the validation sets.

The confusion matrix also appears to be fairly balanced, which is very hopeful.

```python
# testing the model
# Evaluate the model on the test data using `evaluate`
print("Evaluate on test data")
results = model.evaluate(x_test, y_test, batch_size=128)
print("test loss, test acc:", results)
```
